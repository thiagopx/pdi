### Nyquist

Original signal

* Frequency $f_{max}=0{,}5 \text{ Hz}$
* Nyquist rate $R_{Nyquist} = 2 \cdot f_{max} = 1\text{ Hz}$

Sampler 1
* Sampling rate: $R_1 = 2{,}2 \text{ Hz} > R_{Nyquist}$

Sampler 2 (critical sampling)
* Sampling rate: $R_2 = 1 \text{ Hz} = R_{Nyquist}$

Sampler 3 (undersampling)
* Sampling rate: $R_3 = 0{,}6 \text{ Hz} < R_{Nyquist}$

The Nyquist frequency is half of the sampling frequency used to convert continuous signals into discrete samples, and it represents the highest frequency that can be accurately represented in the digital domain.

```python
T = 20.0
f_max = 0.5    # signal frequency
t = np.linspace(0, T, 1000, endpoint=True)
f_t = np.sin(2 * np.pi * f_max * t)

# sampler 1
R1 = 2.2
N1 = int(R1 * T)
t1 = np.linspace(0, T, N1, endpoint=False)
f_t1 = np.sin(2 * np.pi * f_max * t1)

# sampler 2
R2 = 1
N2 = int(R2 * T)
t2 = np.linspace(0, T, N2, endpoint=False)
f_t2 = np.sin(2 * np.pi * f_max * t2)

# sampler 
R3 = 0.6
N3 = int(R3 * T)
t3 = np.linspace(0, T, N3, endpoint=False)
f_t3 = np.sin(2 * np.pi * f_max * t3)

fig, axs = plt.subplots(3, 1, figsize=(12, 9), sharex=True, dpi=80)

for i, (ti, f_ti) in enumerate([(t1, f_t1), (t2, f_t2), (t3, f_t3)]):
    axs[i].plot(t, f_t, label="$f(t)$", color="gray")
    axs[i].set_title(f"sampler {i+1}")
    axs[i].scatter(ti, f_ti, marker="o", label="sampler 2", color="blue")
    axs[i].grid(True)
plt.xlabel("$t$")
plt.show()

print(f_t2)
```
